# Linear Regression

* A model that represent the linear relation between a dependent variable "y" and one or more independing variables "X".
* The model is a line the line equation is shown below
$ y = slope * X + bias $

* We will create synthetic data set, then add some noise to it and use tensorflow to create a model that predicts the correct labels of the data using its noise counterpart.

```python
# Import the libraries we will use

import numpy as np
from matplotlib.pyplot import *
import matplotlib.mlab as mlab
import pandas as pd
import tensorflow as tf
import pandas as pd
from IPython.display import display
```

```python
# === Create data and simulate results ===
# numpy.random.rand generates values from a uniform distribution
# with mean 0 and standard deviation 1

x_data = np.random.rand(2000,3)
features = pd.DataFrame(x_data)
w_real = [0.3,0.5,0.1]
b_real = -0.2

# create a figure
fig1 = figure(num=None, figsize=(20, 4), dpi=80, facecolor='w', edgecolor='k')


# We plot the data. Remember that the probability density
# can be higher than "1", the smaller the sample the higher 
# the data density will be.
fig1_s131 = fig1.add_subplot(subplot(131))
fig1_s131.set_title('FIG 1a - Uniform Distribution')
count, bins, ignored = fig1_s131.hist(x_data, 15, normed=True)
fig1_s131.plot(bins, np.ones_like(bins), linewidth=2, color='r')


# We create some noise and calculate the y_data labels with
# the noise added
noise = np.random.rand(1,2000)*0.1
noise_labels = np.matmul(w_real,features.T) + b_real + noise
real_labels = np.matmul(w_real,features.T) + b_real
features['noise_labels'] = noise_labels.T
features['real_labels'] = real_labels.T

# How the vectors look like
display(features.head(10))

# Plot one column vs its real labels
# Collect samples
samples= features.sample(100)

# Fit a line with the real labels
fit = np.polyfit(samples[0], samples['real_labels'], deg=1)

# Plot one column vs its real labels
fig1_s132 = fig1.add_subplot(subplot(132))
fig1_s132.set_title('FIG 1c - Linear Regression and Real Sample')
fig1_s132.set_xlabel('Column 0')
fig1_s132.set_ylabel('real_labels')
fig1_s132.set_xlim([0.0,1.0])
fig1_s132.set_ylim([0.0,1.0])
fig1_s132.plot(samples[0],fit[0] * samples[0] + fit[1], color='red')
fig1_s132.scatter(samples[0],samples['real_labels'])


# Plot one column vs its noise labels
fig1_s133 = fig1.add_subplot(subplot(133))
fig1_s133.set_title('FIG 1b -Linear Regression and Noise Sample')
fig1_s133.set_xlabel('Column 0')
fig1_s133.set_ylabel('noise_labels')
fig1_s133.set_xlim([0.0,1.0])
fig1_s133.set_ylim([0.0,1.0])
fig1_s133.plot(samples[0],fit[0] * samples[0] + fit[1], color='red')
fig1_s133.scatter(samples[0],samples['noise_labels'],color='green')



# Print all the plots
show()

```

# TensorFlow and Linear Regression

Using tensorflow we will find the optimal weigths for the noise data which will be as close as possible as the real weigths. Tensor flow is finding a linear function that represents the original data .

```python
# INPUT features.iloc[:, :3]
# LABEL features['noise_labels']
# PARAMETERS w_real, b_real
# LABEL prediction Linear Regression
# LOSS function R^2 mean square differece
# OPTIOMIZATION function Gradient Descent

# Number of steps.
# On each epoch the weigths are optimize
NUM_EPOCHS = 200

# Create a non-defualt graph
g = tf.Graph()

# Set a list to hold the regression parameters
weigth_bias_parms = []

# Graph
with g.as_default():
    x = tf.placeholder(tf.float32,shape=[None,3])
    y_true = tf.placeholder(tf.float32,shape=None)
    
    # Define scopes per each stage of the main session
    
    with tf.name_scope('inference') as scope:
        # Create tensorflow weigths variable shape,type,name
        w = tf.Variable([[0,0,0]],dtype=tf.float32, name='weights')
        # Create tensorflow bais variable shape,type,name
        b = tf.Variable(0, dtype=tf.float32, name='bias')
        # Generate the prediction
        y_pred = tf.matmul(w,tf.transpose(x)) + b
   
    with tf.name_scope('loss') as scope:
        loss = tf.reduce_mean(tf.square(y_pred - y_true))
        
    with tf.name_scope('train') as scope:
        learning_rate = 0.5
        optimizer = tf.train.GradientDescentOptimizer(learning_rate)
        train = optimizer.minimize(loss)
        
    # Initialize the Variables by creating its content
    init = tf.global_variables_initializer()
    
    # Start the session
    with tf.Session() as sess:
        sess.run(init)
        print("Iniatl weigths and bias")
        print(sess.run([w,b]),"\n")
        print("Optimized weigths...")
        for epoch in range(NUM_EPOCHS):
            sess.run(train,feed_dict={x: features.iloc[:, :3], y_true: features['noise_labels']})
            # Every 5 epochs print the parameters
            # append this to a list
            if (epoch % 50 == 0):
                print(epoch, sess.run([w,b]))
                weigth_bias_parms.append(sess.run([w,b]))
        
        # Print the optimized weights and bias
        print(200,sess.run([w,b]),"\n")
        
        # Print Original weigths
        print("Original Weigths\n")
        print(w_real, b_real)
                
            
        

```

# Logistic Regression

PROBLEM: A power plant engines fail with no previous warning, the plant engineer has notice that some of the engines sound different from others.

```python
# Load the dataset to a dataframe
data = pd.DataFrame.from_csv('data/engines_data.csv')

# Print data of good and faulty engines
display(data.iloc[0:50:5])

# Plot the data
fig2 = figure(num=None, figsize=(10, 4), dpi=80, facecolor='w', edgecolor='k')
fig2_s111 = fig2.add_subplot(subplot(111))
fig2_s111.set_title('FIG 2 - Generators Dataset Scatter Plot')
fig2_s111.set_xlabel('RPM')
fig2_s111.set_ylabel('Vibration')

for i in range(len(data)):
    x,y,c = data.iloc[i]
    if data['Status'].iloc[i] == 'good':
        fig2_s111.scatter(x,y,color='DarkBlue',marker='^', label='Good')
    else:
        fig2_s111.scatter(x,y,color='DarkGreen',marker='x', label='Faulty')


show()
```

```python
# Create a figure of the scatter plot and the linear regression line
fig3 = figure(num=None, figsize=(10, 5), dpi=80, facecolor='w', edgecolor='k')
fig3_s111 = fig3.add_subplot(subplot(111))
fig3_s111.set_title('FIG 3 - Separation Line')
fig3_s111.set_xlabel('RPM')
fig3_s111.set_ylabel('Vibration')

# We fit a linear regression using the data
fit2 = np.polyfit(data['RPM'],data['Vibration'] , deg=1)
slope = fit2[0]
intercept = fit2[1]
print("slope: ",fit2[0],"intercept: ",fit2[1])
print('''
Note that there is a second slope factor multiplying the Vibration parameter which
it is not shown here; this is important to define the boundary equation. The slope
and the Intercept will be replaced by the parameters w and b.
''')
# We plot both the scatter data and the linear regression
fig3_s111.plot(data['RPM'],slope * data['RPM'] + intercept, color='red')

# Draw the plot
for i in range(len(data)):
    x,y,c = data.iloc[i]
    if data['Status'].iloc[i] == 'good':
        if x == 892:
            fig3_s111.scatter(x,y,color='DarkBlue',marker='^', label='Good')
        else:
            fig3_s111.scatter(x,y,color='DarkBlue',marker='$-$', label='Good')           
    else:
        if x == 722:
            fig3_s111.scatter(x,y,color='DarkGreen',marker='x', label='Faulty')
        else:
            fig3_s111.scatter(x,y,color='DarkGreen',marker='+', label='Faulty')

fig3_s111.annotate('  (810,495) {}'.format(intercept + slope * 810 - 495), xy=(892, 370))
fig3_s111.annotate('  (722,193) {}'.format(intercept + slope * 722 - 193), xy=(722, 193))

show()
```

## The decision boundary equation is equal to
$intercept + slope * RPM = Vibration$
<br>
or
<br>
The boundary is a hyperplane so its equation can be re-written as:
<br>
$intercept + slope * RPM  - Vibration = 0$
<br>
or
<br>
$b + w_1*RPM - w_2*Vibration = 0$
<br>
or
<br>
$b + w_1*x_1 - w_2*x_2 = 0$
<br>
Anything above the decision boundary is "-" and anything below is "+"

```python
#Enter the sigmoid function.
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

fig4 = figure(num=None, figsize=(10, 4), dpi=80, facecolor='w', edgecolor='k')
fig4_s111 = fig4.add_subplot(subplot(111))
fig4_s111.set_title('FIG 4a - Logistic Function')
fig4_s111.set_xlabel('wTx+b')
fig4_s111.set_ylabel('Pr(y=1|x)')

b = intercept
w1 = slope
print("w1:  -0.608289011876 w2: 1.0 b:  784.530832447" )
#print("intercept - slope * RPM - 1 * VIBRATION = 0\n")

# Calculate the value of each point in relation to the 
# Hyperplane
data['logits'] = data.apply(lambda row: sigmoid((b + w1 * row['RPM'] - row['Vibration'])),axis=1)
data['y_linear'] = data.apply(lambda row: b + w1 * row['RPM'] - row['Vibration'], axis=1)
print(data.head(3))
fig4_s111.plot(data['y_linear'],data['logits'],'ro')
show()
```

# Logistic Regression Sample Data

```python
# Define the number of samples

N_logist = 20000

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
    
# Create the sample data

x_data = np.random.randn(N_logist,3)
w_real = [0.3,0.5,0.1]
b_real = -0.2
wxb = np.matmul(w_real,x_data.T) + b_real

y_data_sigmoid = sigmoid(wxb)
y_data_true = np.random.binomial(1,y_data_sigmoid) # Generate data labels


# Plot the sample data
fig5 = figure(num=None, figsize=(15, 4), dpi=80, facecolor='w', edgecolor='k')
fig5_s121 = fig5.add_subplot(subplot(121))
fig5_s121.set_title('FIG 5a - y sigmoid')
fig5_s121.set_xlabel('Wtx + b')
fig5_s121.set_ylabel('Pr(y=1|x)')

fig5_s122 = fig5.add_subplot(subplot(122))
fig5_s122.set_title('FIG 5a - y true')
fig5_s122.set_xlabel('Wtx + b')
fig5_s122.set_ylabel('y')

fig5_s121.plot(wxb,y_data_sigmoid, 'ro')
fig5_s122.plot(wxb,y_data_true, 'ro')

show()

```

# Tensorflow and Logistic Regression

Find the data separation hyperplane using Tensorflow

```python
# INPUT x_data
# LABEL y_data_post_noise
# PARAMETERS w_real, b_real
# LABEL prediction function (single layer or multiple hidden layers)
# LOSS function Cross Entropy
# OPTIOMIZATION function Gradient Descent

NUM_EPOCHS = 50

g = tf.Graph()
weigth_bias_parms = []

with g.as_default():
    x = tf.placeholder(tf.float32,shape=[None,3])
    y_true = tf.placeholder(tf.float32,shape=None)
    
    with tf.name_scope('inference') as scope:
        w = tf.Variable([[0,0,0]],dtype=tf.float32,name='weights')
        b = tf.Variable(0, dtype=tf.float32,name='bias')
        # Label prediction
        y_pred = tf.matmul(w,tf.transpose(x)) + b

    #with tf.name_scope('loss') as scope:
    #    y_pred = tf.sigmoid(y_pred)
    #    loss = y_data_true * tf.log(y_pred) - (1 - y_data_true) * tf.log(1 - y_pred)
    #    loss = tf.reduce_mean(loss)
        
    with tf.name_scope('loss') as scope:
        loss = tf.nn.sigmoid_cross_entropy_with_logits(labels=y_true,logits=y_pred)
        loss = tf.reduce_mean(loss)
    
    with tf.name_scope('train') as scope:
        learning_rate = 0.5
        optimizer = tf.train.GradientDescentOptimizer(learning_rate)
        train = optimizer.minimize(loss)
        
    init = tf.global_variables_initializer()
    with tf.Session() as sess:
        sess.run(init)
        for epoch in range(NUM_EPOCHS):
            sess.run(train, feed_dict={x: x_data, y_true: y_data_true})
            if (epoch % 5 == 0):
                print(epoch,sess.run([w,b]))
                weigth_bias_parms.append([w,b])
                
        print(50, sess.run([w,b]))
        print(w_real,b_real)
```
